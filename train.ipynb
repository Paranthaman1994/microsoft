```python
# Apply the method described
threshold = len(df) * 0.5
df_cleaned = df.dropna(axis=1, thresh=threshold)
removed_columns = df.columns[df.isnull().sum() > threshold]


# print(df_cleaned)
```

```python
df_cleaned["IncidentGrade"] = df_cleaned["IncidentGrade"].fillna(df_cleaned["IncidentGrade"].mode().iloc[0])
```

```python
df_cleaned.info()
```

```python
df1=df_cleaned
```

```python
df1.columns = df1.columns.str.lower().str.replace(' ', '_')
```

```python
df1
```

```python
column_to_move = 'incidentgrade'

# Create a list of columns with the specified column moved to the end
columns = [col for col in df1.columns if col != column_to_move] + [column_to_move]

# Reorder the DataFrame columns
df1 = df1[columns]
```

```python
df1
```

```python
categorical_columns = df1.select_dtypes(include=['object', 'category'])
```

```python
categorical_columns
```

```python
# Specify categorical columns for label encoding
categorical_columns = ['timestamp', 'category', 'entitytype', 'evidencerole','incidentgrade']  # Replace with your actual categorical columns

# Initialize the LabelEncoder
label_encoder = LabelEncoder()

# Perform label encoding for each categorical column
for column in categorical_columns:
    if column in df1.columns:
        # df1[column] = label_encoder.fit_transform(df1[column].astype(str))
        df1[column] = label_encoder.inverse_transform(df1[column].astype(str))

```
